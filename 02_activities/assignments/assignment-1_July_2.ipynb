```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Creating a simple test plot using random values for y
np.random.seed(613) # this is needed for reproductability 
x = np.arange(50)
y = np.random.randint(0,100,50)

fig, ax = plt.subplots(figsize=(5,3)) 
ax.scatter(x,y) # this is a scatter plot
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.bar(x,y) # this is a bar plot
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.hist(y) # this is a hist plot
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

#putting somemore details into our visualization 
ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
```

```python
#setting styles for the font
font1 = {'family': 'sans-serif', 'color': 'blue', 'size': 20}
font2 = {'family': 'monospace', 'color': 'green', 'size': 14}

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total growth over time', fontdict = font1, loc = 'left')
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x, y, marker = '*',  # chaning style of the plot
           color = 'indigo')

fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x, y, marker = '*',
           color = 'indigo',
           linestyle = '--',
           linewidth = 2)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x, y, marker = '*',
           color = '#742589',
           linestyle = '--',
           linewidth = 2,
           markersize = 12,
           markeredgecolor = '#FA9359',
           markerfacecolor = '#000000')

ax.grid(color = 'blue', linewidth = 1, linestyle = '-') #adding the grid for better visual
```

```python
#Activity 2

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
# Data
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x) + np.random.uniform(size=len(x)) - 0.2
# Create a color if the y axis value is equal or greater than 0
my_color = np.where(y>=0, 'green', 'skyblue')
# The vertical plot is made using the vline function
plt.vlines(x=x, ymin=0, ymax=y, color=my_color, alpha=0.4)
plt.scatter(x, y, color=my_color, s=1, alpha=1)

# Add title and axis names
plt.title("Evolution of the value of x", loc='left')
plt.xlabel('Value of the variable')
plt.ylabel('Group')
# Show the graph
plt.show()

#It is aesthetic, seems to be substentive and perceptual.
```
